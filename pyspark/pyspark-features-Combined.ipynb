Examples Of Pyspark ML

```python
## Read The dataset
training = spark.read.csv('/drive/My Drive/Colab Notebooks/pyspark/BasicCommands/data/test1.csv',header=True,inferSchema=True)
```

```python
training.show()
```

```python
training.printSchema()
```

```python
training.columns
```

[Age,Experience]----> new feature--->independent feature

```python
from pyspark.ml.feature import VectorAssembler
featureassembler=VectorAssembler(inputCols=["age","Experience"],outputCol="Independent Features")
```

```python
output=featureassembler.transform(training)
```

```python
output.show()
```

```python
output.columns
```

```python
finalized_data=output.select("Independent Features","Salary")
```

```python
finalized_data.show()
```

```python
from pyspark.ml.regression import LinearRegression
##train test split
train_data,test_data=finalized_data.randomSplit([0.75,0.25])
regressor=LinearRegression(featuresCol='Independent Features', labelCol='Salary')
regressor=regressor.fit(train_data)
```

```python
### Coefficients
regressor.coefficients
```

```python
### Intercepts
regressor.intercept
```

```python
### Prediction
pred_results=regressor.evaluate(test_data)
```

```python
pred_results.predictions.show()
```

```python
pred_results.meanAbsoluteError,pred_results.meanSquaredError
```

Overview
This notebook will show you how to create and query a table or DataFrame that you uploaded to DBFS. DBFS is a Databricks File System that allows you to store data for querying inside of Databricks. This notebook assumes that you have a file already inside of DBFS that you would like to read from.

This notebook is written in Python so the default cell type is Python. However, you can use different languages by using the %LANGUAGE syntax. Python, Scala, SQL, and R are all supported.

```python
# File location and type
file_location = "/drive/My Drive/Colab Notebooks/pyspark/BasicCommands/data/tips.csv"
file_type = "csv"

# The applied options are for CSV files. For other file types, these will be ignored.
df =spark.read.csv(file_location,header=True,inferSchema=True)
df.show()
```

```python
df.printSchema()
```

```python
df.columns
```

```python
### Handling Categorical Features
from pyspark.ml.feature import StringIndexer
```

```python
indexer=StringIndexer(inputCol="sex",outputCol="sex_indexed")
df_r=indexer.fit(df).transform(df)
df_r.show()
```

```python
indexer=StringIndexer(inputCols=["smoker","day","time"],outputCols=["smoker_indexed","day_indexed",
                                                                  "time_index"])
df_r=indexer.fit(df_r).transform(df_r)
df_r.show()
```

```python
df_r.columns
```

```python
from pyspark.ml.feature import VectorAssembler
featureassembler=VectorAssembler(inputCols=['tip','size','sex_indexed','smoker_indexed','day_indexed',
                          'time_index'],outputCol="Independent Features")
output=featureassembler.transform(df_r)
```

```python
output.select('Independent Features').show()
```

```python
output.show()
```

```python
finalized_data=output.select("Independent Features","total_bill")
```

```python
finalized_data.show()
```

```python
from pyspark.ml.regression import LinearRegression
##train test split
train_data,test_data=finalized_data.randomSplit([0.75,0.25])
regressor=LinearRegression(featuresCol='Independent Features', labelCol='total_bill')
regressor=regressor.fit(train_data)
```

```python
regressor.coefficients
```

```python
regressor.intercept
```

```python
### Predictions
pred_results=regressor.evaluate(test_data)
```

```python
## Final comparison
pred_results.predictions.show()
```

```python
### PErformance Metrics
pred_results.r2,pred_results.meanAbsoluteError,pred_results.meanSquaredError
```
